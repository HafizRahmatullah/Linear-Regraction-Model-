```python
import numpy as np
import pandas as pn
import matplotlib.pyplot as plt
```

```python
data = pn.read_csv('/home/rahmat/Desktop/Data csv file for ML/LRMV.csv')
data.head()
```

```python
x = data[['area','bedrooms','age']].values
y = data['price'].values
```

```python
plt.scatter(data['area'],y)
plt.show()
```

```python
x0 = np.ones(len(x))
theta = np.array([0,0,0,0])
X = np.array([x0,data['area'],data['bedrooms'],data['age']]).T
x_transpose = X.T
x_inverse = np.linalg.pinv(x_transpose.dot(X))
x_transpose_y = x_transpose.dot(y)
normal_equaction = x_inverse.dot(x_transpose_y)
plt.scatter(data['area'], y)
plt.plot(data['area'],X.dot(normal_equaction) , label='Plots the line predicted by the normal equation solution ')
plt.title("Level vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)

plt.scatter(x[:,0],y)
plt.plot(x[:,0],y)
plt.show()
```

```python
model.predict(np.array(x))
```
